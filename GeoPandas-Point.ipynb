```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
# Read point data
data = gpd.read_file('PointData.shp')
```

```python
# Define the projected CRS
data = data.to_crs('EPSG:32618')
```

```python
data.head()
```

# X and Y Coordinates: Extract the longitude (x-coordinate) and latitude (y-coordinate) values from each point in the dataset.

```python
# Extract x and y coordinates
data['x'] = data.geometry.x
data['y'] = data.geometry.y
```

```python
# View the coordinates
data.head()
```

# Distance Metrics: Calculate distances between points

```python
#Specify index point to which we need to find distance from
index_point = Point(292178.489, 39347.601)

# Calculate the distance between each row and the index point
distances = data.geometry.distance(index_point)
```

```python
# Add the distances as a new column in the GeoDataFrame
data['distance_to_index'] = distances
```
